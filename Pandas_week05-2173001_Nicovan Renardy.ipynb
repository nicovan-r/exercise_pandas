```python
import pandas as pd
import numpy as np
```

```python
n_rows = 365 * 24 #Membentuk data deret waktu yang interval hourly basis mengcover durasi satu tahun. 24 merepresentasikan jumlah jam dalam satu hari, 365 merepresentasikan jml hari dalam satu tahun
n_cols = 2
cols = ['col1', 'col2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), #mengisi data frame dengan bilangan integer random dalam jangka 1 sampai 20
                  columns=cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')#mengisi index data deret tanggal dan waktu internalnya waktu per jam berdasarkan n_rows
df
```

```python
df.resample('M')['col1'].sum().to_frame()#proses pengelompokan ulang data berdasarkan interval waktu tertentu. df.resample('M') artinya per bulan
#fungsi sum mengubah dari berdasarkan jam ke data bulan menimbulkan agregasi, data tak bisa diakses per individual data 
#.to_frame() untuk membuat tampilan frame tabel.
```

```python
df.resample('D')['col1'].sum().to_frame()#proses pengelompokan ulang data berdasarkan interval waktu tertentu. df.resample('D') artinya per hari
#hasil total jml baris ada 365 karena satu tahun ada 365 hari
```

```python
n_rows = 5
n_cols = 2
cols = ['omset', 'operasional']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

```python
df['omset'] = df['omset'] * 100_000 #100_00 untuk mengkonversi nilai ke satuan rupiah 100 ribuan. Underscore di Python sebagai ignore character untuk digital placeholder
df['operasional'] = df['operasional'] * 10_000 #100_00 untuk mengkonversi nilai ke satuan rupiah 100 ribuan
df
```

```python
df.index = pd.util.testing.makeDateIndex(n_rows, freq='D')#membuat kolom baru tanggal untuk indexnya
df = df.reset_index()#artinya membuat kolom bertipe tanggal bukan index nya tanggal. reset_index() untuk mengubah index saat ini ke kolom biasa bernama index
df = df.rename(columns={'index':'tanggal'})
df
```

```python
#membuat dictionary penampung format tiap kolom, urutan kolom dapat beda sesuai dengan urutan di formatku. Pengformatan bertipe data String makanya diapiit petik
formatku = {'tanggal':'{:%d/%m/%y}', #mengubah format tanggal jadi date month year
            'operasional':'Rp {:.2f}', 
            'omset':'Rp {:.2f}'} #menambahkan penanda rupiah (Rp). {:.2f} menambahkan dua digit extra di belakang koma bernilai .00

laporan = df.style.format(formatku)
laporan #variabel laporan bukan bertipe data data frame, tetapi pandas.io.formats.style.Styler jadi ga bisa pakai fungsi yang ada di data frame
```
