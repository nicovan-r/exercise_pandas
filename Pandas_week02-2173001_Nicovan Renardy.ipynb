```python
df = pd.util.testing.makeMissingDataframe().reset_index() 
df.head()
#.reset_index() untuk menambahkan index yang dari sananya ke nomor urut dari no 0
```

```python
df = df.rename(columns={'index':'Z'})
df.head()
```

```python
df_backup = df.copy(deep=True)
#membackup data
```

```python
df = df.dropna(axis='columns') 
df.head()
```

```python
df = df_backup.copy(deep=True)
df = df.dropna(axis='rows')
df.head()
```

```python
df = df_backup.copy(deep=True)
df.isna().mean()
#menghitung rata2 missing values setiap kolom
```

```python
treshold = len(df) * 0.9 #thresold= ambang batas tertentu, len(df)*0.9 = mengetahui 90% dari jumlah rows yg missing values di dataframe
df = df.dropna(thresh=treshold, axis='columns')
df.head()
```
