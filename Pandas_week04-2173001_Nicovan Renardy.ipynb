```python
import pandas as pd
import numpy as np
```

```python
n_rows = 10
n_cols = 1
cols = ('usia',)

df = pd.DataFrame(np.random.randint(1, 99, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

```python
df['kelompok_usia'] = pd.cut(df['usia'], 
                             bins=[0, 18, 65, 99], #bins bertipe List dan bernilai integer dengan koma sebagai pembatasnya. usia 0 sampai 18:  anak, 18 sampai 65: dewasa, 65-99: manula
                             labels=['anak', 'dewasa', 'manula'])#isi kelompok_usia menggunakan variable labels
df
```

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

```python
df1 = df.copy(deep=True)
df1 = df1.drop([1, 4])#membuang isi index 1 dan 4 dari data frame df
df1
```

```python
df2 = df.copy(deep=True)
df2 = df2.drop([0, 3])#membuang isi index 0 dan 3 dari data frame df
df2
```

```python
df_inner = pd.merge(df1, df2, how='inner')#how adalah cara pendekatan. Inner mengambil isi baris data frame yang dimiliki kedua df yang terpecah
df_inner#merge akan memproduksi index yang baru
```

```python
df_outer = pd.merge(df1, df2, how='outer')#how adalah cara pendekatan. Outer mengambil isi baris data frame yang tidak sama punya dari kedua df yang terpecah
df_outer#merge akan memproduksi index yang baru
```

```python
data = {'nama':['Didi Kempot', 'Glenn Fredly', 'Mbah Surip'], 
        'tempat_kelahiran':['Surakarta, Jawa Tengah', 'Jakarta, DKI Jakarta', 'Mojokerto, Jawa Timur']}#data frame dibentuk dari suatu dictionary yang terdiri dari dua buah i artinya ada 2 buah kolom: nama dan tempat kelahiran
df = pd.DataFrame(data)
df
```

```python
df[['nama_depan', 'nama_belakang']] = df['nama'].str.split(' ', expand=True)#split untuk pemecah string berdasarkan suatu karakter. expand = true artinya mengizinkan output berupa kolom baru bernama nama depan dan nama belakang
df
```

```python
df[['kota', 'propinsi']] = df['tempat_kelahiran'].str.split(',', expand=True)
df
```

```python
df = pd.read_csv('./data/titanicfull.csv')#titanicfull.csv adalah sample data frame untuk keperluan data science, survived: 1 survived dan 0 not survived
df.head()
```

```python
df.groupby(['sex', 'pclass'])['survived'].mean().to_frame()#groupby mengelompokan data berdasarkan kelompok index sex dan pclass. Oleh karena itu, tak dapat mengakses raw nilai dari survived, bisanya aggregate value dari data survived sekarang berupa rata2.
#.to_frame() untuk mempercantik data tertampil dalam bentuk frame
```

```python
df.groupby(['sex', 'pclass'])['survived'].mean().unstack()#reshaping dengan fungsi .unstack(). Kolom survived ga ada, langsung nilai survivednya saat dipresentasikan
```
