Pandas 41: Menerapkan agregasi pada sejumlah kolom dengan agg()

```python
df = pd.read_csv('./data/titanicfull.csv')
df.head()
```

```python
df.groupby('pclass').agg({'pclass':'count', #kali ini fungsi agg() menerima suatu tipe data dictionary
                          'age':['mean', 'max'], #key menjadi nama kolom yang dilakukan fungsi agregasi dan value menandakan fungsi agregasi yang diterapkan pada kolom tersebut. Jika ada lebih dr satu fungsi agregasi, maka perlu dimasukan ke dalam list.
                          'survived': 'mean'})
```

```python
df.groupby('pclass').agg(n_pass=('pclass', 'count'),#menggunakan judul kolom tertentu yang lebih representatifkan fungsi yang digunakan. fungsi agg menggunakan keyword argument agg(pemberian nama kolom yang lebih spesifik=('nama kolom yang diberi fungsi', 'fungsi')
                         avg_age=('age', 'mean'),
                         max_age=('age', 'max'), 
                         survival_rate=('survived', 'mean'))
```

Pandas 43: Menangani Whitespace pada data frame

```python
data = {'nim': ['10', '11', '12', '13', '  '],
        'nama': ['adi', '  ', 'tejo', '  ', 'bejo']}#isi index dengan spasi saja akan menimbulkan masalah pada data frame berupa tidak terdeteksinya whitespace sebagai null atau NaN

df = pd.DataFrame(data)
df
```

```python
df
```

```python
df.info()#menampilkan informasi mengenai data frame
```

```python
df = df.replace(r'^\s*$', np.nan, regex=True)#solusi ingin mendeteksi kemunculan karakter spasi murni tidak disertai karakter lain menggunakan methode replace dengan regex atau regular expression diset sebagai True lalu regex pattern selalu dimulai dengan karakter 'r'dan pattern dari regex. Pattern: '^\s*$' ^ menunjukan diawali, $ diakhiri, \s karakter spasi tunggal, * menandakan keberadaan karakter \s atau sederet spasi yang jumlahnya tidak menentu
df#np.nan merepresentasikan missing value
```

```python
df.info()
```

Pandas 44: Menata ulang penempatan kolom pada data frame

```python
n_rows = 5#jumlah baris
n_cols = 5#jumlah kolom
cols = tuple('ABCDE')#nama kolom A,B,C,D,E

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)), 
                  columns=cols)#randint jangkauan nilai dari 1 sampai 10
df
```

```python
#Menata ulang penempatan kolom pada data frame
df[['D', 'C', 'A', 'E', 'B']]#penataan ulang kolom/rearrange nama kolom[[kolom 1, kolom 2, ...]], sifatnya tidak menggantikan data frame baru melainkan membuat data frame baru dengan urutan seperti ini
```
