```python
import pandas as pd
import numpy as np
```

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

```python
df.shape #hasil: (10, 5) artinya jumlah 10 baris dan 5 kolom
```

```python
proporsi = 0.7 #70% untuk data frame pertama, sisa 30% untuk data frame kedua
df_1 = df.sample(frac=proporsi) #membuat bagian data frame acak 70% yang diadopsi dari data diatas. df adalah data frame original
df_2 = df.drop(df_1.index) #drop untuk menghapus sebagian isi data frame sesuai data baris2 yang sudah diadopsi dari df 70% 

print(f'df_1 shape: {df_1.shape}')
print(f'df_2 shape: {df_2.shape}')
```

```python
df.loc[df['B']>10, ['B','D','E']] #kalo istilahnya di SQL, df['B']>10 sebagai where dan ['B','D''E] sebagai select
```

```python
df.loc[0:4, 'B':'D'] #Beda slicing pada List adalah pada loc star index:end index bersifat inklusif semua atau diturut sertakan. Pada slicing list, start index inklusif dan end index eksklusif. 
```

```python
data = {'day':[1, 2, 10 ,25, 12], 
        'month':[1, 2, 4, 5, 6], 
        'year':[2000, 2001, 2010, 2015, 2020]} #data ini berstruktur data dictionary. dan harus day, month, year dll dengan bahasa inggris 

df = pd.DataFrame(data)
df
```

```python
df['penaggalan'] = pd.to_datetime(df[['day', 'month', 'year']]) #harus menggunakan nama kolom day month year dan dibungkus ke dalam bentuk list
df#membuat kolom dataframe berbentuk datetime dari kolom integre di atas
```
