Pandas 46: Seleksi weekdays dan weekends pada data deret waktu (time series)

```python
#persiapan data frame
n_rows = 365#banyak baris
n_cols = 2#banyak kolom
cols = ['col1', 'col2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq='D')#membentuk data deret waktu denganpd.util.testing.makeDateIndex(jumlah baris, freq='D')
df
```

```python
#seleksi weekdays
weekdays_df = df[df.index.dayofweek.isin([0, 1, 2, 3, 4])]#df.index.dayofweek bisa diakses dengan catatan data framenya time series/index datetime. index.dayofweek().isin([index hari kerja]). isin untuk mengecek apakah dayofweeknya ada pada deret angka yang dimasukan. 0 senin, 1 selasa, 2 rabu, ...
weekdays_df.head(7)#menampilkan 7 baris pertama dari weekdays
```

```python
#seleksi weekends
weekends_df = df[df.index.dayofweek.isin([5, 6])]#df.index.dayofweek bisa diakses dengan catatan data framenya time series/index datetime. index.dayofweek().isin([index hari akhir pekan]). isin untuk mengecek apakah dayofweeknya ada pada deret angka yang dimasukan. 5 sabtu, 6 minggu
weekends_df.head(7)
```

Pandas 48: Mengenal cumulative count dengan cumcount() 

```python
#data frame penjualan barang
d = {'penjual':['bejo', 'tejo', 'wati', 'bejo', 'cecep', 'tejo', 'wati', 'bejo'], #salesperson
     'barang':['monitor', 'monitor', 'keyboard', 'mouse', 'keyboard', 'monitor', 'laptop', 'monitor']}#barang yang dijual

df = pd.DataFrame(d)
df
```

```python
#cumcount() untuk mengkumulatifkan count atau jumlah baris atau kemunculan nilai
df['count_tiap_penjual'] = df.groupby('penjual').cumcount() + 1#menyertakan plus 1 agar cumcount()nya mulai dari satu bukan nol
df
```

```python
df['count_tiap_barang'] = df.groupby('barang').cumcount() + 1
df
```

```python
df['count_pasangan_kolom'] = df.groupby(['penjual', 'barang']).cumcount() + 1#dikelompokan berdasarkan kolom penjual dan barang
df
```
