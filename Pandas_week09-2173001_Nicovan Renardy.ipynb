Pandas 33: Menampilkan nilai kumulatif (cumulative)

```python
#Persiapan Data Frame
d = {'pemain':['Budi', 'Joni', 'Iwan', 'Budi', 'Budi', 'Iwan', 'Asep', 'Joni'], #kolom pemain terdiri dari nama-nama pemain sepak bola yang berhasil mencetak goal
     'goal':[2, 1, 3, 1, 1, 2, 2, 3]}#kolom goal berisi jumlah goal yang berhasil dicetak oleh seorang pemain dalam suatu sesi pertandingan.
df = pd.DataFrame(d)
df
```

```python
#Menampilkan nilai kumulatif
df['goal'].cumsum().to_frame()#fungsi cumulative sum yang berfungsi mendata jumlah goal secara kumulatif, jumlah goal saat ini akan dijumlahkan dengan keseluruhan goal sebelumnya
```

```python
df['jumlah_goal_kumulatif'] = df['goal'].cumsum()#Kolom baru jumlah goal kumulatif, menampung cumulative sum dari goal
df
```

```python
df['jumlah_goal_kumulatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum()#menerapkan cumulative sum untuk goal berdasarkan nama seorang pemain
df
```

```python
df['cummax'] = df['goal'].cummax()#fungsi cumulative max, mendata nilai tertinggi sampai dengan titik tertentu pada goal
df
```

```python
df['cummin'] = df['goal'].cummin()#fungsi cumulative min, mendata nilai terendah sampai dengan titik tertentu pada goal
df
```

```python
df['cumprod'] = df['goal'].cumprod()#fungsi cumulative product, hasil total perkalian kumulatif
df
```

Pandas 34: Mapping pada Data Frame dengan applymap()

```python
#Persiapan Data Frame
df = pd.DataFrame({'jenis_kelamin':['Pria', 'Wanita', 'lelaki', 'Lelaki', 'perempuan'], #Isi data frame kolom jenis_kelamin dan shift tidak konsisten, yang selanjutnya akan distandarisasi dengan mapping
                   'usia':[23, 21, 24, 22, 21], 
                   'shift':['pagi', 'siang', 'Malam', 'Siang', 'pagi']})
df
#Isi data frame kolom jenis_kelamin dan shift tidak konsisten
```

```python
#Melakukan mapping dengan applymap()
df = df.applymap(lambda x: x.lower() if type(x) == str else x)#lambda expression dikenakan keseluruhan data frame df menerima nilai x lalu diuji tipe datanya, jika tipe string maka nilai string jadi lowercase semua tapi jika bukan akan dikembalikan seperti semula
df
```

```python
#dictionary ini melakukan mapping untuk standarisasi terhadap data-data pada Data Frame
mapping = {'pria':'L',
           'lelaki':'L',
           'wanita':'P',
           'perempuan':'P',
           'pagi':1,
           'siang':2,
           'malam':3}#jika seperti ini mappingnya, maka akan kena ke seluruh kolom Data Frame

df.applymap(mapping.get) #applymap(variabel dictionary mapping)
```

```python
df[['jenis_kelamin', 'shift']] = df[['jenis_kelamin', 'shift']].applymap(mapping.get)#melakukan mapping ke kolom tertentu. df[['nama kolom','nama kolom']] = df[['nama kolom','nama kolom']]

df
```

Pandas 34: Memadukan agregasi dengan fungsi transform()

```python
#Persiapan Data Frame Transaksi Coffee Shop
d = {'no_nota':[1, 1, 1, 2, 2, 3, 4, 5], 
     'kopi': ['latte', 'cappuccino', 'espresso', 'latte', 'espresso', 'cappuccino', 'latte', 'espresso'],
     'harga':[50, 60, 80, 150, 120, 60, 100, 40]}

df = pd.DataFrame(d)
df
```

```python
#Menampilkan total harga untuk tiap nomor nota
df.groupby('no_nota')['harga'].sum().to_frame()#pengelompokan data-data berdasarkan no_nota dan dikenakan fungsi agregasi sum()
```

```python
df['total_harga'] = df.groupby('no_nota')['harga'].transform(sum)#kolom baru total_harga berisi data-data total gabungan no_nota yang sama dengan transform(sum)
df
```

```python
#Menghitung total omset untuk tiap jenis kopi yang terjual
df.groupby('kopi')['harga'].sum().to_frame()#mengelompokan Data Frame berdasarkan kopi, lalu menerapkan fungsi sum() pada kolom harga
```

```python
df['total_omset'] = df.groupby('kopi')['harga'].transform(sum)#kolom baru total_omset berisi data-data total gabungan harga kopi yang sama dengan transform(sum)
df
```

Pandas 36: Menyatukan kolom dengan str.cat()

```python
#Persiapan Data Frame
data = {'nama': ['bayu', 'indra', 'devi', 'agni'],
        'jenis_kelamin': ['L', 'L', 'P', 'L'], 
        'usia': [23, 21, 22, 25]}

df = pd.DataFrame(data)
df
```

```python
#Menyatukan kolom dengan str.cat()
df['nama'].str.cat(df['jenis_kelamin'], sep=', ').to_frame()#Menggabungkan nilai kedua kolom yang dipisah (sep) dengan , koma selama dua-dua nilainya bertipe data string
```

```python
df['nama_jk'] = df['nama'].str.cat(df['jenis_kelamin'], sep=', ')
df
```

```python
df['nama'].str.cat(df['usia'].astype('str'), sep=' - ').to_frame()#menggabungkan isi kolom yang bertipe data String dan Integer. Konversikan integer ke dalam tipe data string dengan astype('str')
```

```python
df['nama_usia'] = df['nama'].str.cat(df['usia'].astype('str'), sep=' - ')# Tidak perlu menggunakan to_frame() karena data ditampung ke kolom data frame
df
```
