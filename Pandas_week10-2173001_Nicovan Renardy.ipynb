Pandas 37: Seleksi baris dengan banyak kriteria

```python
#persiapan Data Frame
df = pd.read_csv('./data/titanicfull.csv')
df.head()
```

```python
#Seleksi baris dengan banyak kriteria
df[(df['sex']=='female') & (df['age']>=60) & (df['embarked']=='S') & (df['survived']==1)]#setiap kriteria harus diapit dengan kurung buka dan kurung tutup dan kriteria-kriteria dirangkai dengan tanda '&'
```

```python
#meningkatkan reability dari code dengan merangkaikan tiap kriteria pada baris terpisah
df[
    (df['sex']=='female') & 
    (df['age']>=60) & 
    (df['embarked']=='S') & 
    (df['survived']==1)
]
```

```python
#meningkatkan reability dari code dengan memasukan kriteria ke dalam variabel sendiri masing-masing
kr1 = df['sex']=='female'
kr2 = df['age']>=60
kr3 = df['embarked']=='S'
kr4 = df['survived']==1

df[kr1 & kr2 & kr3 & kr4]#melakukan seleksi baris
```

Pandas 38: Mengenal parameter header dan skiprows

header untuk menetapkan suatu row menjadi header, digunakan ketika ada row yang seharusnya berada di bagian paling atas atau index ke 0 tetapi bukan berada di index ke 0. Proses header mengakibatkan index-index sebelum dari index yang ditentukan menjadi header terlewati.
skiprows untuk melewati row yang tidak ingin ditampilkan

```python
#persiapan data frame
df = pd.read_csv('./data/iris_error.csv')
df.head(8)
```

```python
df = pd.read_csv('./data/iris_error.csv', header=2, skiprows=[5,6]) #Pertama read_csv dulu yang memuat parameter berisi (direktori file csv berada, header=index ke berapa di file csv yang ingin menjadi header, skiprows=indeks mana saja yang ingin dilewati)
df.head()
```

Pandas 39: Mengacak urutan baris pada data frame

```python
n_rows = 6
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

```python
df.sample(frac=1.0, random_state=1)#melakukan pengacakan kemunculan barisnya (shuffle process) dengan method sample(frac=1.0, random_state=1). frac=1.0 artinya melakukan sampling dengan proporsi yang penuh setiap entry data/ index akan dimunculkan secara acak sesuai random_state number, random_state number dispesifikasikan untuk menjamin agar apa yang dilakukan bisa diraplikasi
```

```python
df.sample(frac=1.0, random_state=1).reset_index(drop=True) #mengembalikan urutan index semula dengan menambahkan .reset_index(drop=True)
```

Pandas 40: Mengakses sekelompok data dengan get_group()

```python
df = pd.read_csv('./data/titanicfull.csv')
df.head()
```

```python
#Mengakses sekelompok data yang sudah terkelompok dengan get_group()
grouped_df = df.groupby('sex')#titanic dataset dikelompokan berdasarkan sex yang menampung dua kelompok data female dan male
```

```python
grouped_df.get_group('female').head(10)#mengakses group dengan jenis sex female saja. head(10) artinya mengakses 10 baris pertama saja
```

```python
grouped_df = df.groupby('survived')#titanic dataset dikelompokan berdasarkan survive status yang menampung dua kelompok data 0 dan 1
```

```python
grouped_df.get_group(1).head(10)#mengakses group dengan survived bernilai 1 saja. head(10) artinya mengakses 10 baris pertama saja
```
